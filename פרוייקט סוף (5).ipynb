```python
! pip install seaborn
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
df = pd.read_csv('https://raw.githubusercontent.com/verticale3185/FINEL-PROJECT/main/cristiano_vs_messi.csv')
df
```

```python
df.assisted = df.assisted.fillna('Solo')
df.date = pd.to_datetime(df.date)
df['year'] = pd.DatetimeIndex(df['date']).year
messi_df = df.loc[df.player == 'messi']
ronaldo_df = df.loc[df.player == 'ronaldo']
ronaldo_df
```

```python
if len(messi_df.pos) > 2:
    messi_df.pos = messi_df.pos.str[0:2]
messi_df.groupby(['pos', 'player'])[['pos']].count()
```

```python
messi_df.groupby(['pos', 'player']).sum().plot(kind='pie', subplots=True, shadow = True, startangle=90,figsize=(6, 6))
```

```python
if len(ronaldo_df.pos) > 2:
    ronaldo_df.pos = ronaldo_df.pos.str[0:2]
ronaldo_df.groupby(['pos', 'player'])[['pos']].count()
```

```python
ronaldo_df.groupby(['pos', 'player']).sum().plot(kind='pie', subplots=True, shadow = True, startangle=90,figsize=(3,3))
```
